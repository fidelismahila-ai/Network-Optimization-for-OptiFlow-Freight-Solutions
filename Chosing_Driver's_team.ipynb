## Solve Optimization for Team A

### Subtask:
Using Gurobi, formulate and solve an optimization problem for Team A. This will involve defining decision variables for the total hours each Team A driver works over the entire trip and the number of days the trip takes. Constraints will include individual driver daily hour limits, the total team's 18-hour daily limit, and the total distance (3805 miles) to be covered. The objective will be to minimize Team A's total cost. After solving, record Team A's minimum cost, total driving time, and number of days.


**Reasoning**:
I will import the Gurobi library and define the optimization model for Team A, including decision variables, constraints, and the objective function as outlined in the instructions. After optimizing, I will extract and store the relevant results.



```python
# Install Gurobi
!pip install gurobipy

# Import the Gurobi Package
import gurobipy as gp
from gurobipy import GRB

# Create a new model
m = gp.Model("TeamA_Optimization")

# Define decision variables
x_pro_A = m.addVar(lb=0, vtype=GRB.CONTINUOUS, name="x_pro_A") # Total hours for Pro driver A
x_amateur_A = m.addVar(lb=0, vtype=GRB.CONTINUOUS, name="x_amateur_A") # Total hours for Amateur driver A
num_days_A = m.addVar(lb=0, vtype=GRB.CONTINUOUS, name="num_days_A") # Total number of days for Team A

# Constraints
# Individual driver daily hour limits
m.addConstr(x_pro_A <= team_A_data['Pro']['hours_per_day'] * num_days_A, "Pro_A_Daily_Limit")
m.addConstr(x_amateur_A <= team_A_data['Amateur']['hours_per_day'] * num_days_A, "Amateur_A_Daily_Limit")

# Team daily driving constraint
m.addConstr(x_pro_A + x_amateur_A <= team_driving_constraint * num_days_A, "Team_A_Daily_Limit")

# Total distance covered
m.addConstr(x_pro_A * team_A_data['Pro']['speed'] + \
            x_amateur_A * team_A_data['Amateur']['speed'] >= distance_to_cover, "Total_Distance_A")

# Set objective function: Minimize total cost
m.setObjective(x_pro_A * team_A_data['Pro']['rate'] + \
               x_amateur_A * team_A_data['Amateur']['rate'], GRB.MINIMIZE)

# Optimize the model
m.optimize()

# Retrieve and store results
if m.status == GRB.OPTIMAL:
    team_a_cost = m.objVal
    team_a_time = x_pro_A.x + x_amateur_A.x
    team_a_num_days = num_days_A.x
    print(f"Team A Optimal Cost: {team_a_cost:.2f}")
    print(f"Team A Total Driving Time: {team_a_time:.2f} hours")
    print(f"Team A Number of Days: {team_a_num_days:.2f}")
else:
    print("No optimal solution found for Team A")
    team_a_cost = float('inf')
    team_a_time = float('inf')
    team_a_num_days = float('inf')
```
